```python
import yaml, pandas as pd
config=yaml.safe_load(open('configs/training_config.yaml',encoding='utf-8'))
daily=pd.read_pickle(config['dataset']['output'])
```

```python
from src.expression import Expression
expressions=[Expression.generate(max_depth=4,seed=i) for i in range(10)]
for expr in expressions:
    values=expr.execute(daily)
    print(str(expr), 'complexity=',expr.complexity(),'depth=',expr.depth(),'coverage=',round(values.notna().mean(),3))
```
